```python
xtrain_data = data.loc[:1800, :]
x=xtrain_data.drop(['price_range'], axis=1)
y=xtrain_data['price_range']

xtest_data = data.loc[1800:, :]
x2=xtest_data.drop(['price_range'], axis=1)
y2=xtest_data['price_range']
```

# Decition Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(min_samples_leaf=8)
model.fit(x, y)
#max_depth
#min_samples_leaf
model.score(x,y)
```

```python
model.score(x2,y2)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(n_estimators=100,min_samples_leaf=2)
model2.fit(x, y)
#max_depth
#min_samples_leaf
print(model2.score(x,y))
```

```python
model2.score(x2,y2)
```

# Nearest Neighbors  #KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier(n_neighbors=20)
model3.fit(x, y)
#max_depth
#min_samples_leaf
print(model3.score(x,y))
```

```python
model3.score(x2,y2)
```
